# Tâche 1
Ce notebook trace les fonctions \( f(x) = x^2 \) et \( g(x) = 2x \).

```python
import numpy as np
import matplotlib.pyplot as plt

# Créer un tableau NumPy avec 20 éléments entre -1 et 1
x = np.linspace(-1, 1, 20)

# Définir les fonctions
def f(x):
    return x**2

def g(x):
    return 2 * x

# Tracer les fonctions
plt.plot(x, f(x), label="f(x) = x^2")
plt.plot(x, g(x), label="g(x) = 2x")

# Ajouter des labels et une légende
plt.xlabel("x")
plt.ylabel("y")
plt.title("Graphique de f(x) = x^2 et g(x) = 2x")
plt.legend()
plt.grid()
plt.show()
```
